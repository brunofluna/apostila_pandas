## Resumo sobre PANDAS

---

PANDAS
- Pandas  uma Biblioteca do Python voltada para a estruturao de dados;
- Permite a leitura e escrita de dados em vrios formatos: CSVs, Excel, SQL, entre outros;
- Permite a realizao de anlise de dados e modelagem;

---

<font color='red'>Pontos positivos</font>
. Extremamente fcil de usar, exigindo uma pequena curva de aprendizagem para lidar com dados tabulares;
. Conjunto grande de ferramentas para carregar, transformar e escrever dados em vrios formatos;
. Principal escolha para a maioria das bibliotecas usadas em Machine Learning;
. Capacidade para grficos e visualizaes de dados (usando matplotlib).

<font color='red'>Pontos negativos</font>
. Grande ocupao da memria, pois cria muitos objetos adicionais para rpido acesso e facilidade de manipulao;
